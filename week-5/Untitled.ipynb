```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 2*pi with a small step
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate corresponding y values for the sine function
y = np.sin(x)

# Plot the sine function
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')

# Display the plot
plt.show()
```
