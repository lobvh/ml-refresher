```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH = './imgs'
NAME = 'leena.jpg'

FULL_NAME = f'{PATH}/{NAME}'
```

```python
leena_img = load_img(FULL_NAME, target_size = (300, 400))
```

```python
leena_img
```

```python
# Convert the image to a NumPy array
leena_array = img_to_array(leena_img).astype(np.uint8)

# Convert the image to grayscale
gray_leena = cv2.cvtColor(leena_array, cv2.COLOR_RGB2GRAY)

# Apply the Sobel filter for vertical edges
sobel_filter = np.array([[-1, 0, 1],\
                         [-2, 0, 2],\
                         [-1, 0, 1]])
sobel_result = cv2.filter2D(gray_leena, -1, sobel_filter)

# Display the original and Sobel-filtered images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(leena_array.astype(np.uint8))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Sobel Filtered Image')
plt.imshow(sobel_result, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# Apply the Edge Detection filter 
edge_filter = np.array([[-1, -1, -1],\
                         [-1, 8, -1],\
                         [-1, -1, -1]])
edge_result = cv2.filter2D(gray_leena, -1, edge_filter)

# Display the original and Sobel-filtered images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(leena_array.astype(np.uint8))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edge Detected Image')
plt.imshow(edge_result, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# Create a white 30x30 image
image = np.ones((30, 30), dtype=np.uint8) * 255

# Add a vertical black line
image[:, 15] = 0

# Apply Sobel filter for edge detection
sobel_filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
edge_result = cv2.filter2D(image, -1, sobel_filter)

# Display the original image, the vertical line, and the Sobel edge result
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axvline(x=15, color='red', linestyle='--', linewidth=2)
plt.title('Vertical Black Line')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edge_result, cmap='gray', vmin=0, vmax=255)
plt.title('Sobel Edge Detection Result')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Create a white 30x30 image
image = np.ones((30, 30), dtype=np.uint8) * 255

# Add a vertical black line
image[:, 15] = 0

# Define the edge enhancement filter
edge_filter = np.array([[-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])

# Apply the filter using convolution
edge_result = cv2.filter2D(image, -1, edge_filter)

# Display the original image, the vertical line, and the edge enhancement result
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axvline(x=15, color='red', linestyle='--', linewidth=2)
plt.title('Vertical Black Line')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edge_result, cmap='gray', vmin=0, vmax=255)
plt.title('Edge Enhancement Result')
plt.axis('off')

plt.show()
```

```python
# Create a white 30x30 image
image = np.ones((30, 30), dtype=np.uint8) * 255

# Add a vertical black line
image[:, 15] = 0

# Create a 5x5 averaging filter
blur_filter = np.ones((5, 5), dtype=np.float32) / 25.0

# Apply the averaging filter for smoothing
smoothed_image = cv2.filter2D(image, -1, blur_filter)

# Display the original, line, and smoothed images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axvline(x=15, color='red', linestyle='--', linewidth=2)
plt.title('Vertical Black Line')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(smoothed_image, cmap='gray', vmin=0, vmax=255)
plt.title('Smoothed Image (Averaging Filter)')
plt.axis('off')

plt.show()
```
