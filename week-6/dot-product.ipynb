```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector(ax, v, color, label):
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
    ax.set_xlim([-1, 4])
    ax.set_ylim([-1, 4])
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(color='gray', linestyle='--', linewidth=0.5)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

def main():
    # Define 2D vectors
    v = np.array([1, 3], dtype=float)  # Convert to float
    w = np.array([2, -1], dtype=float)  # Convert to float

    # Calculate dot product
    dot_product = np.dot(v, w)

    # Calculate angle between vectors
    cos_theta = dot_product / (np.linalg.norm(v) * np.linalg.norm(w))
    theta = np.arccos(cos_theta) * (180 / np.pi)

    # Create a 2D plot
    fig, ax = plt.subplots()

    # Plot vector v
    plot_vector(ax, v, 'r', 'Vector v')

    # Plot vector w
    plot_vector(ax, w, 'b', 'Vector w')

    # Plot the line perpendicular to v
    v_perpendicular = np.array([-v[1], v[0]], dtype=float)  # Convert to float
    v_perpendicular /= np.linalg.norm(v_perpendicular)  # Make it a unit vector

    # Draw the line
    x_vals = np.linspace(-1, 4, 100)
    y_vals = (v_perpendicular[1] / v_perpendicular[0]) * x_vals
    ax.plot(x_vals, y_vals, linestyle='dashed', color='g', label='Line Perpendicular to v')

    # Print results
    print(f"Vector v: {v}")
    print(f"Vector w: {w}")
    print(f"Dot Product (v ⋅ w): {dot_product}")
    print(f"Cosine of the Angle (cos θ): {cos_theta}")
    print(f"Angle (θ) in degrees: {theta}°")

    # Show the plot
    ax.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
