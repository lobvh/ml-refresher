# Linear Scale

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
def draw_uniform_distribution():
    # Create figure and axis
    fig, ax = plt.subplots()

    # Draw the uniform distribution (rectangle from 0 to 100)
    uniform_rect = patches.Rectangle((0, 0), 100, 1, linewidth=1, edgecolor='b', facecolor='b', alpha=0.2)
    ax.add_patch(uniform_rect)

    # Define ranges to be highlighted
    ranges = [(0.0001, 0.001, 'r', 0.4), (0.001, 0.01, 'g', 0.4), (10, 100, 'y', 0.4)]

    # Draw shaded rectangles for specified ranges
    for range_start, range_end, color, alpha in ranges:
        range_rect = patches.Rectangle((range_start, 0), range_end - range_start, 1, linewidth=1, edgecolor=color, facecolor=color, alpha=alpha)
        ax.add_patch(range_rect)

    # Set axis limits and labels
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 1)
    ax.set_xlabel('Value')
    ax.set_ylabel('Probability Density')

    # Show the plot
    plt.show()

# Call the function to draw the plot
draw_uniform_distribution()
```

# Log Scale

```python
# Set the range for the log-uniform distribution
min_value = 0.0001
max_value = 100

# Generate a large number of samples from the log-uniform distribution
num_samples = 1000
log_uniform_samples = np.power(10, np.random.uniform(np.log10(min_value), np.log10(max_value), num_samples))

# Define the intervals
interval1 = [0.0001, 0.001]
interval2 = [0.001, 0.01]
interval3 = [10, 100]

# Plot a histogram to visualize the distribution
plt.hist(log_uniform_samples, bins=50, edgecolor='black', alpha=0.7)
plt.xscale('log')  # Set x-axis to logarithmic scale for better visualization

# Highlight the intervals
plt.axvspan(interval1[0], interval1[1], color='red', alpha=0.3, label=f'{interval1}')
plt.axvspan(interval2[0], interval2[1], color='green', alpha=0.3, label=f'{interval2}')
plt.axvspan(interval3[0], interval3[1], color='blue', alpha=0.3, label=f'{interval3}')

plt.title('Log-Uniform Distribution')
plt.xlabel('Value')
plt.legend()

plt.show()
```

```python
# Defining samples for both scales

def uniform_vs_loguniform(lower_bound, upper_bound, num_samples):
    #Linear sampling
    linear_samples = list(np.linspace(lower_bound, upper_bound, num_samples))

    #Logarithmic sampling
    log_samples = list(10 ** np.linspace(np.log10(lower_bound), np.log10(upper_bound), num_samples))
    
    print(f"----Samples between {lower_bound} and {upper_bound}----")
    print("Linear Samples:", linear_samples)
    print("Logarithmic Samples:", log_samples)
```
