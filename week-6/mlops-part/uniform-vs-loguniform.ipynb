```python
# Set the range for the log-uniform distribution
min_value = 0.0001
max_value = 100

# Generate a large number of samples from the log-uniform distribution
num_samples = 100000
log_uniform_samples = np.power(10, np.random.uniform(np.log10(min_value), np.log10(max_value), num_samples))

# Define the intervals
interval1 = [0.0001, 0.001]
interval2 = [0.001, 0.01]
interval3 = [10, 100]

# Count the samples in each interval
count_interval1 = np.sum((log_uniform_samples >= interval1[0]) & (log_uniform_samples < interval1[1]))
count_interval2 = np.sum((log_uniform_samples >= interval2[0]) & (log_uniform_samples < interval2[1]))
count_interval3 = np.sum((log_uniform_samples >= interval3[0]) & (log_uniform_samples < interval3[1]))

# Print the counts
print(f"Number of samples in {interval1}: {count_interval1}")
print(f"Number of samples in {interval2}: {count_interval2}")
print(f"Number of samples in {interval3}: {count_interval3}")

# Plot a histogram to visualize the distribution
plt.hist(log_uniform_samples, bins=50, edgecolor='black', alpha=0.7)
plt.xscale('log')  # Set x-axis to logarithmic scale for better visualization

# Highlight the intervals
plt.axvspan(interval1[0], interval1[1], color='red', alpha=0.3, label=f'{interval1}')
plt.axvspan(interval2[0], interval2[1], color='green', alpha=0.3, label=f'{interval2}')
plt.axvspan(interval3[0], interval3[1], color='blue', alpha=0.3, label=f'{interval3}')

plt.title('Log-Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

Let's break down the intervals in the logarithmic space:

[0.0001, 0.001] (log scale):

log10(0.0001) = -4
log10(0.001) = -3
The length of this interval in the log scale: -3 - (-4) = 1

[0.001, 0.01] (log scale):

log10(0.001) = -3
log10(0.01) = -2
The length of this interval in the log scale: -2 - (-3) = 1

[10, 100] (log scale):

log10(10) = 1
log10(100) = 2
The length of this interval in the log scale: 2 - 1 = 1

```python
# Set the range for the uniform distribution
min_value = 0.0001
max_value = 100

# Generate a large number of samples from the uniform distribution
num_samples = 100000
uniform_samples = np.random.uniform(min_value, max_value, num_samples)

# Define the intervals
interval1 = [0.0001, 0.001]
interval2 = [0.001, 0.01]
interval3 = [10, 100]

# Count the samples in each interval
count_interval1 = np.sum((uniform_samples >= interval1[0]) & (uniform_samples < interval1[1]))
count_interval2 = np.sum((uniform_samples >= interval2[0]) & (uniform_samples < interval2[1]))
count_interval3 = np.sum((uniform_samples >= interval3[0]) & (uniform_samples < interval3[1]))

# Print the counts
print(f"Number of samples in {interval1}: {count_interval1}")
print(f"Number of samples in {interval2}: {count_interval2}")
print(f"Number of samples in {interval3}: {count_interval3}")

# Plot a histogram to visualize the distribution
plt.hist(uniform_samples, bins=50, edgecolor='black', alpha=0.7)

# Highlight the intervals
plt.axvspan(interval1[0], interval1[1], color='red', alpha=0.6, label=f'{interval1}')
plt.axvspan(interval2[0], interval2[1], color='green', alpha=1, label=f'{interval2}')
plt.axvspan(interval3[0], interval3[1], color='blue', alpha=0.3, label=f'{interval3}')

plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
